# Analysis for PD Subtype

```python
import sys
sys.path.append('/mnt/0A2AAC152AABFBB7/CGE/luxgiant-clinical')
import luxgiant_clinical.ClinicalReport as lxc

import pandas as pd
```

The following analysis are done to PD patients for whom it is known the disease subtype.

```python
# read local data
df = pd.read_csv('/mnt/0A2AAC152AABFBB7/CGE/luxgiant-clinical/data/cleaned_file.csv', low_memory=False)

# filters
mask_patients= df['Status']=='Patient'
mask_subtype = ~df['subtype'].isnull()

# filter data
df_cases = df[mask_patients & mask_subtype].reset_index(drop=True)

df_cases
```

Basic descriptive statistics are show for the studied samples.

```python
variables_4_patients = [
    'age', 'age_at_onset', 'bmi_comp', 'total_score_for_moca', 'PD_duration', 'years_of_education', 'total_score_for_bdi'
]
lxc.descriptive_analytics(df_cases, variables=variables_4_patients)
```

```python
cond = 'OFF: Off is the typical functional state when patien ts have a poor response in spite of taking medications.'
lxc.conditioned_descriptive_analytics_by_group(
    df_data=df_cases,
    variables=variables_4_patients,
    group_var='subtype',
    condition=('on_off', cond)
)
```

```python
cond = 'OFF: Off is the typical functional state when patien ts have a poor response in spite of taking medications.'
lxc.oneway_anova(
    df_data  =df_cases,
    variables=variables_4_patients,
    group_var='subtype',
    condition=('on_off', cond)
)
```

```python
cond = 'OFF: Off is the typical functional state when patien ts have a poor response in spite of taking medications.'
lxc.multi_variable_bonferroni(
    df_data=df_cases,
    variables=variables_4_patients,
    group_var='subtype',
    condition=('on_off', cond)
)
```

```python
cond = 'OFF: Off is the typical functional state when patien ts have a poor response in spite of taking medications.'
lxc.multi_variable_kw(
    df_data=df_cases,
    variables=variables_4_patients,
    group_var='subtype',
    condition=('on_off', cond)
)
```

```python
crosstab_vars = [
    'zone_of_origin', 'education_level', 'agecat_1', 'agecat_2', 'pdsl', 'handedness', 'work', 'reason', 
    'nature_of_work___1', 'nature_of_work___2', 'nature_of_work___3', 'over_your_lifetime_have_yo', 
    'during_your_lifetime_did_y', 'in_your_lifetime_have_you', 'in_your_lifetime_have', 
    'have_you_ever_had_a_head_i', 'l_dopa_carbidopa', 'pramipexole', 'ropinerole', 'rasageline', 
    'selegeline', 'entacapone', 'amantadine', 'apomorphine', 'medical_history_neurologic___7', 
    'medical_history_cancer___1', 'medical_history_cancer___2', 'medical_history_cancer___3', 
    'medical_history_cancer___4', 'medical_history_metabolic___1', 'medical_history_metabolic___3', 
    'medical_history_metabolic___4', 'medical_history_cardiovasc___1', 'medical_history_cardiovasc___2', 
    'medical_history_cardiovasc___3', 'medical_history_psychologi___1', 'medical_history_psychologi___3', 
    'family_member_diagnosed_wi', 'family_member_diagnosed_wi_2', 'family_member_diagnosed', 
    'ct_head', 'mri_brain_imaging', 'datscan', 'f_dopa_pet', 'pd_surgical_treatment', 'dbs_done'
]
```

```python
lxc.chi_squared_analysis(
    df_data             =df_cases,
    conditionants       =[('on_off', cond)],
    variables_4_analysis=crosstab_vars,
    grouping_var        ='subtype'
)
```

```python
crosstab_vars_2 = [
    'handedness', 'nature_of_work___2', 'nature_of_work___3', 'selegeline', 'apomorphine', 'medical_history_neurologic___7', 
    'medical_history_cancer___1', 'medical_history_cancer___2', 'medical_history_cancer___3', 'family_member_diagnosed'
]
lxc.chi_squared_analysis(
    df_data             =df_cases,
    conditionants       =[('on_off', cond)],
    variables_4_analysis=crosstab_vars_2,
    grouping_var        ='subtype'
)
```

```python
lxc.intergroup_kw_with_bonferroni_correction_one_variable(
    df_data=df_cases,
    variable='age',
    group_var='subtype'
)
```

```python
lxc.multi_variable_kw_bonferroni(
    df_data=df_cases,
    variables=variables_4_patients,
    group_var='subtype',
    condition=('on_off', cond)
)
```

```python
conditionants = [
    [('on_off', cond), ('ct_head', 'Yes')], [('on_off', cond), ('mri_brain_imaging', 'Yes')],
    [('on_off', cond), ('datscan', 'Yes')], [('on_off', cond), ('f_dopa_pet', 'Yes')]
]
var_conds = [['ct_result___1'], ['mri_result___1'], ['datscan_result___1'], ['f_dopa_pet_result___1']]

df_res = pd.DataFrame()

for k in range(len(conditionants)):

    df_temp = lxc.chi_squared_analysis(
        df_data             =df_cases,
        conditionants       =conditionants[k],
        variables_4_analysis=var_conds[k],
        grouping_var        ='subtype'
    )
    df_res = pd.concat([df_res, df_temp], axis=0, ignore_index=True)

df_res
```
